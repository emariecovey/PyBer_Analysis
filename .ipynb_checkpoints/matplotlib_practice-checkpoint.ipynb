```python
#magic command that makes sure plots are displayed in Jupyter Notebook right after code cell that produced it
%matplotlib inline

#importing matplotlib dependency
import matplotlib.pyplot as plt
import statistics
import numpy as np

#to add ticker marks between x axis number tickers in horizontal bar charts
from matplotlib.ticker import MultipleLocator
```

```python
savings = [100,300,500,900,700]
months = ['jan','feb','mar','apr','may']

#plots (x,y) and x, y have to have same number in them
plt.plot(months, savings)
plt.xlabel("Months")
plt.ylabel("Savings in $")
plt.title("Savings Account Results Per Month")


#only shows most recent plot
#plt.show() should only be used once per session because it can cause the graph to display improperly. If you have problems with data not being displayed properly, you may have to restart your kernel and clear your output.
plt.show()
```

```python
### CREATING A LINE CHART USING MATLAB

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#find standard deviation of y_axis
stdev=statistics.stdev(y_axis)

#set figure size (before making plot) (width and then height)
plt.figure(figsize=(20,10))

#make plot and label for legend. Changed line color and width and markers on line
plt.plot(x_axis, y_axis, label="Boston", marker="*", color="blue", linewidth=2)

#make x and y axis labels 
plt.xlabel("Date")
plt.ylabel("Fares ($)")
#set y axis limits
plt.ylim(0,45)
#setting x axit limits so it only shows until month of june
plt.xlim("Jan","June")
#title on chart
plt.title("PyBer Fare by Month")
#making legend
plt.legend()
#adding grid lines to chart
plt.grid()

```

```python
### CREATING LINE CHART USING OBJECT-ORIENTED INTERFACE

#The plt.subplots() function returns a tuple that contains a figure and axes object(s).
#fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
#The first variable, fig, references the figure.
#The second variable, ax, references the axes of the figure and allows us to annotate the figure.
#The ax.plot(x_axis, y_axis) plots the data from the x and y axes.

#The fig, ax = plt.subplots() line of code can also be written like this:
#fig = plt.figure()
#ax = fig.add_subplot()
#ax.plot(x_axis, y_axis)

#to see a quick figure with object-oriented code:
#ax = plt.axes()
#ax.plot(x_axis, y_axis)

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="d", linewidth=2, label="Boston")

#displays label "Boston"
ax.legend()
#x and y axis labels, title
ax.set_xlabel("Date")
ax.set_ylabel("Fares ($)")
ax.set_title("Pyber Fare Per Month")
#add y axis limit
ax.set_ylim(0,45)
ax.grid()
```

```python
#MAKING LINE CHART WITH ERROR BARS

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#find standard deviation of y_axis
stdev=statistics.stdev(y_axis)

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#MAKING LINE CHARTS WITH ERROR BARS OBJECT-ORIENTED APPROACH

fig,ax=plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#CREATING VERTICAL BAR CHART WITH MATLAB

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#make the bar chart
plt.bar(x_axis, y_axis, color="green", label="Boston", yerr=stdev, capsize=3)
#displays legend
plt.legend()
#adding labels and title
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare By Month")
```

```python
#CREATING HORIZONTAL BAR CHART WITH MATLAB
#good for graphing nominal data or data where it's sorted greatest to least

#creates plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
#inverts the plot so that january is at top, not december
plt.gca().invert_yaxis()
#makes ticks every $5 instead of every $10 (default)
plt.xticks(np.arange(0,51,step=5.0))
#displaying legend
plt.legend()
#adding labels and title
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare By Month")
```

```python
### CREATING VERTICAL BAR CHART USING OBJECT-ORIENTED INTERFACE

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="New York", yerr=stdev, capsize=3)
#making title and axis lables 
ax.set_xlabel("Month")
ax.set_ylabel("Fares ($)")
ax.set_title("Pyber Fare By Month")
#display legend
ax.legend()
```

```python
### CREATING HORIZONTAL BAR CHART USING OBJECT-ORIENTED INTERFACE

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
#display legend
ax.legend()
#x and y axis labels, title
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Month")
ax.set_title("Pyber Fare By Month")
#switching order of months so jan is first
ax.invert_yaxis()
#setting x ticks to be every 5 instead of 10
ax.set_xticks(np.arange(0,51,step=5.0)) #could also set y ticks if desired #have to import numpy first
#making ticks between x ticks
ax.xaxis.set_minor_locator(MultipleLocator(1)) #have to import multiple locator first 
```

```python
#CREATING SCATTERPLOT WITH MATLAB

#can do it with .plot and "o" or .scatter
#plt.plot(x_axis, y_axis, "o")

#making y axis values larger so that we can see the data points better, since they're set to be the size of the marker
larger_values = []

for y in y_axis:
    larger_values.append(y*3)
#then you could put s = larger values, or just put the line [i*3 for i in y_axis], same thing

plt.scatter(x_axis, y_axis, color="red", label="Chicago", s=larger_values) #s is size of marker. can also set it equal to a number (like 20) or variable
#table and axis labels, legend
plt.xlabel("Month")
plt.ylabel("Fares ($)")
plt.title("Pyber Fare By Month")
plt.legend()
#plt.gca().invert_yaxis()

```

```python
#CREATING SCATTERPLOT WITH OBJECT-ORIENTED INTERFACE

#to do:
1. #Change the color of the markers to sky blue. (c="skyblue")
2. #Change the size of the markers to 5 times each data point. (s=[i*5 for i in y_axis])
3. #Make the color 20% transparent. (alpha=.8)
4. #Add a black edge color to the circles.(edgecolor="black")
5. #Make the linewidth of the circles 2 points. (linewidth=2)
6. #Add a legend for the city of Boston. (label=boston, ax.legend())
7. #Add a title. (.set_title)
8. #Switch the axis so that the Fare($) data is on the x-axis. 
9. #Add a limit for the x-axis data. (xlim)
10. #Invert the y-axis so that January is at the top. (invert_yaxis)

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="skyblue", label="Boston", edgecolor="black", linewidth=2, alpha=.8)
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Month")
ax.set_title("Pyber Fare By Month")
ax.legend()
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
#PIE CHARTS WITH MATLAB

plt.subplots(figsize=(10,10))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, autopct="%1.1f%%", explode=explode_values, colors=colors)
plt.show
```

```python
#CREATING PIE CHART WITH OBJECT-ORIENTED INTERFACE

1. #Add a percentage to one decimal place to each wedge of the pie.
2. ####Increase the figure size to 8x8.
3. #Explode the two highest percentage months.
4. #Add a shadow.
5. #Add a start angle so that January is at the top.
6. #Reverse the order so that the month order is in a clockwise direction.
7. #Add new colors of your own choosing or use the colors from the previous pie chart.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, autopct="%1.1f%%", explode=explode_values, colors=colors, shadow=True, startangle=90, radius=2.5, counterclock=False)

plt.show

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.scatter
#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html
```
